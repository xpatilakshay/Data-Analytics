```python
import matplotlib.pyplot as plt 
```

### Line Plot

```python
x = [90,80,50] 
y = [80,110,33]

plt.plot(x,y)
plt.title("Title")
plt.xlabel("X axix")
plt.ylabel("y axis")
plt.show()
```

### Bar Chart

```python
grades = ['A','B','C']
marks = [90,80,70]
plt.bar(grades,marks)
plt.title("Grades VS Marks")
plt.xlabel("Grades")
plt.ylabel("Marks")
plt.show()
```

### Play with Fontsize , bar colors , width

grades = ['A','B','C']
marks = [90,80,70]
plt.bar(grades,marks,width=0.5,color="Black")
plt.title("Grades VS Marks",fontsize=20)
plt.xlabel("Grades",fontsize=15)
plt.ylabel("Marks",fontsize=15)
plt.show()

### Each Bar with Different Colors

```python
grades = ['A','B','C']
marks = [90,80,70]
colors=['RED','BLACK','ORANGE']
plt.bar(grades,marks,width=0.5,color=colors)
plt.title("Grades VS Marks")
plt.xlabel("Grades")
plt.ylabel("Marks")
plt.show()
```

### Changing Bar Edge Color
alpha : used to control visibility lie between 0 to 1
linewidth : used to control width of an edge

```python
grades = ['A','B','C']
marks = [90,80,70]
colors=['RED','BLACK','ORANGE']
c = ['black','yellow','red']
plt.bar(grades,marks,edgecolor=c,linewidth=5,alpha=0.5,color=colors)
plt.title("Grades VS Marks")
plt.xlabel("Grades")
plt.ylabel("Marks")
plt.show()
```

### creating multiple bar graphs in the single chart

```python
categories = ['A','B','C']
values1 = [10,20,30]
values2 = [40,50,60]


plt.bar(categories,values2,width=0.3,color='Red',label='data2')
plt.bar(categories,values1,width=0.3,color='Black',label='data1')
plt.title("Data")
plt.legend(loc='upper center')

plt.xlabel('Grades')
plt.ylabel('Values')
plt.show()
```

```python
import numpy as np
categories = ['A','B','C']
values1 = [10,20,30]
values2 = [40,50,60]

width = 0.2
g1 = np.arange(len(categories))
g2 = [i+width for i in g1]

plt.bar(g1,values2,width=width,label='data2')
plt.bar(g2,values1,width=width,label='data1')
plt.title("Data")
plt.legend(loc='upper center')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### plt.xticks() is used to:
Set the position and labels of ticks on the x-axis.

```python
import numpy as np
categories = ['A','B','C']
values1 = [10,20,30]
values2 = [40,50,60]

width = 0.2
g1 = np.arange(len(categories))
g2 = [i+width for i in g1]

plt.bar(g1,values2,width=width,label='data2')
plt.bar(g2,values1,width=width,label='data1')
plt.title("Data")
plt.legend(loc='upper center')

plt.xlabel('Categories')
plt.ylabel('Values')


plt.xticks(g1+width/2,categories)
plt.show()
```

### plt.yticks() sets:

Tick positions (where ticks appear on the y-axis)

Tick labels (what text shows at those positions)

```python
import numpy as np
categories = ['A','B','C']
values1 = [10,20,30]
values2 = [40,50,60]

width = 0.2
g1 = np.arange(len(categories))
g2 = [i+width for i in g1]

plt.bar(g1,values2,width=width,label='data2')
plt.bar(g2,values1,width=width,label='data1')
plt.title("Data")
plt.legend(loc='upper center')

plt.xlabel('Categories')
plt.ylabel('Values')


plt.yticks(values1,['Low','Mid','High'])
plt.show()
```

### Scatter plot

```python
x = [10,20,30,40]
y = [100,200,300,400]

plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Scatter plot Example")

plt.show()
```

### some points
s : used to set size od points
color : sets color to points
alpha : set opacity 

```python
x = [10,25,67,78,65,97]
y = [33,4,5,67,67,54]

plt.scatter(x,y,color="red",s=100,label="Data Values",alpha=1)
plt.title("Scatter plot Customization")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Histogram

```python
data = [5,25,35,67,45,68]
plt.hist(data,bins = 5,color="Orange",edgecolor="Black")
plt.title("Histogram")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

### Pie Chart
Show proportions or percentages:

```python
subjects = ['Marathi','Hindi','English','History']
marks  = [97,66,75,91]

plt.pie(marks,labels=subjects,autopct='%1.2f%%',startangle=90)

plt.title("Pie chart")
plt.show()
```

### Pie chart detailed

```python
cars = ['ScorpioN','Thar','BE6','Harrier','Safari']
pollution = [10,8,1,12,14]


plt.pie(pollution,
         labels=cars,
         autopct="%1.2f%%",
         startangle=90,
         explode=[0,0,0.1,0,0],
         shadow=True
        )
plt.savefig("PieChart.png",dpi=300)
plt.title("Pie Chart")
plt.show()
```
